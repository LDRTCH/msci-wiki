# Data Type Comparison

### Import Libraries

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import cmcrameri as cmc

import sys
sys.path.append('./..')
from src.utils import get_ds_iters


```

### Plot Data Types

```python
hf = h5py.File(f"../data/no-rolling/dataset_tumble_0.050_density_0.25.h5", "r")
iters = get_ds_iters(hf.keys())
img = hf[f"conf_{iters[300]}"]
img = np.array(img)

monochrome_img = np.copy(img)
monochrome_img[monochrome_img>0] = 1
scrambled_img = monochrome_img * np.random.randint(1,5,size=(128,128)) / 4
orientation_img = img / 4

monochrome_img = monochrome_img.reshape((img.shape[0], img.shape[1], 1))
scrambled_img = scrambled_img.reshape((img.shape[0], img.shape[1], 1))
orientation_img = orientation_img.reshape((img.shape[0], img.shape[1], 1))


fig,(ax1,ax2,ax3) = plt.subplots(nrows=1,ncols=3,figsize=(10,4),dpi=600)
ax1.matshow(orientation_img,cmap="cmc.lajolla")
ax2.matshow(scrambled_img,cmap="cmc.lajolla")
ax3.matshow(monochrome_img,cmap="cmc.lajolla")
```
